```python
# k, position of the token in input sequence
# n, user defined scaler, in the attention is all you need paper, it's 100k
# d, positional encoding dimension
# P(k, i), where i is the location of positional dim
def pos_encoding(seq_len, d, n = 100000):
    p = np.zeros((seq_len, d))
    for k in range(seq_len):
        for i in np.arange(d//2):
            denom = np.power(n, 2 * i / d)
            p[k, 2*i] = np.sin(k / denom)
            p[k, 2*i + 1] = np.cos(k / denom)
    return p

p = pos_encoding(seq_len= 100, d = 512 , n = 10000)

cax = plt.matshow(p)
plt.gcf().colorbar(cax)
```
