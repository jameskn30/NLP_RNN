```python
from bs4 import BeautifulSoup
import os
import shutil
```

```python
# Load the HTML file

for html_file in os.listdir('./data'):

    yc, ss_year = html_file.split("_")
    ss, ext = ss_year.split(".")
    year = ss[1:]

    if ss[0] == 's':
        season = 'spring'
    elif ss[0] == 'w':
        season = 'winter'
    else:
        season = 'fall'

    with open(f'data/{html_file}', 'r') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    exclude_links = [
        'https://www.ycombinator.com/companies/founders',
    ]

    # Extract all the links that start with 'https://www.ycombinator.com/companies/'
    links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].startswith('https://www.ycombinator.com/companies/') and a['href'] not in exclude_links]
    print(f'season = {ss}, year = {year}')
    print('total = ', len(links))
    print('first 5 links: ')
    for link in links[:5]:
        print(link)
    
    if os.path.exists("output") == False:
        os.makedirs('output')

    with open(f'./output/yc_{season}_{year}.txt', 'w') as file:
        for link in links:
            file.writelines(link + '\n')
    
    print("####")
    
```

```python
# Make a copy in yc_scraper

shutil.copytree('output', '../yc_scraper/yc_scraper/spiders/output')
```
